# Maintain consistant identifiers and unique obs/vars names

* Check the existence of basic dataset identifier and add it if missing.
* Validate the uniqueness of variable names and make them unique if necessary.
* Validate the uniqueness of cell barcodes locally and globally + make them unique if necessary.


```python
def add_basic_dataset_labels(adatas_dict):
    dataset_mappings = {
        'adata_149614': 'GSE149614',
        'adata_151530': 'GSE151530',
        'adata_125449': 'GSE125449',
        'adata_124395': 'GSE124395',
        'adata_115469': 'GSE115469',
        'adata_140228': 'GSE140228',
        'adata_146409': 'GSE146409',
        'adata_136103': 'GSE136103',
        'adata_138709': 'GSE138709',        
    }
    
    for var_name, dataset_name in dataset_mappings.items():
        if var_name not in adatas_dict:
            continue
            
        # Work directly with the dictionary reference
        adata = adatas_dict[var_name]
        
        # Add dataset identifier
        if 'dataset' in adata.obs.columns:
            print(f"'dataset' column already exists in {dataset_name} with value {adata.obs['dataset'].iloc[0]}. Passed.")
        else:
            print(f"Adding 'dataset' column for {dataset_name}")
            adata.obs['dataset'] = dataset_name

        # Validate var names are unique per dataset
        if not adata.var_names.is_unique:
            print(f"Variable names not unique in {dataset_name}. Making them unique.")
            adata.var_names_make_unique()
        else:
            print(f"Variable names are already unique in {dataset_name}.")

        # Validate cell barcodes are unique locally
        if not adata.obs_names.is_unique or any(adata.obs_names.str.startswith('AAACCTGA')):
            print(f"Making cell names unique for {dataset_name}")
            adata.obs_names = dataset_name + '_' + adata.obs_names
            adata.obs_names_make_unique()
        else:
            print(f"Cell names are already unique for {dataset_name}.")   
        
        print(f"Labeled {dataset_name}: {adata.shape}")
        print('='*80)
        
        adatas_dict[var_name] = adata
    
    return adatas_dict

adatas_dict = add_basic_dataset_labels(adatas_dict)
```

```python
def validate_cell_barcodes_global_uniqueness(adatas_dict: Dict, verbose: bool = True) -> Dict:
    validation_report = {
        'total_cells': 0,
        'unique_barcodes': 0,
        'duplicates_found': False,
        'duplicate_details': [],
        'per_dataset_stats': {}
    }
    
    # Collect all cell barcodes across datasets
    all_barcodes = []
    dataset_barcode_map = {}
    
    print("\nSTEP 1: COLLECTING BARCODES FROM ALL DATASETS")
    print("-" * 50)
    
    for name, adata in adatas_dict.items():
        barcodes = adata.obs_names.tolist()
        all_barcodes.extend(barcodes)
        dataset_barcode_map[name] = set(barcodes)
        
        validation_report['per_dataset_stats'][name] = {
            'n_cells': len(barcodes),
            'unique_within_dataset': adata.obs_names.is_unique,
            'n_unique': len(set(barcodes))
        }
        
        if verbose:
            print(f"  {name}:")
            print(f"    Total cells: {len(barcodes):,}")
            print(f"    Unique within dataset: {adata.obs_names.is_unique}")
    
    validation_report['total_cells'] = len(all_barcodes)
    validation_report['unique_barcodes'] = len(set(all_barcodes))
    
    # Check for global duplicates
    print("\nSTEP 2: CHECKING FOR GLOBAL DUPLICATES")
    print("-" * 50)
    
    barcode_counts = pd.Series(all_barcodes).value_counts()
    duplicate_barcodes = barcode_counts[barcode_counts > 1]
    
    if len(duplicate_barcodes) > 0:
        validation_report['duplicates_found'] = True
        print(f"  DUPLICATES FOUND: {len(duplicate_barcodes):,} barcodes appear multiple times")
        print(f"  Total affected cells: {duplicate_barcodes.sum():,}")
        
        # Show top duplicates
        print(f"\n  Top 10 duplicate barcodes:")
        for barcode, count in duplicate_barcodes.head(10).items():
            datasets_with_barcode = [name for name, barcodes in dataset_barcode_map.items() 
                                    if barcode in barcodes]
            
            validation_report['duplicate_details'].append({
                'barcode': barcode,
                'count': int(count),
                'datasets': datasets_with_barcode
            })
            
            print(f"    '{barcode}': appears {count} times in {', '.join(datasets_with_barcode)}")
        
        if len(duplicate_barcodes) > 10:
            print(f"    ... and {len(duplicate_barcodes) - 10:,} more duplicate barcodes")
        
        # Cross-dataset conflicts
        print(f"\n  Cross-dataset conflicts:")
        for name1, barcodes1 in dataset_barcode_map.items():
            for name2, barcodes2 in dataset_barcode_map.items():
                if name1 >= name2:
                    continue
                
                overlap = barcodes1.intersection(barcodes2)
                if len(overlap) > 0:
                    print(f"    {name1} <-> {name2}: {len(overlap):,} shared barcodes")
    else:
        print(f"  NO DUPLICATES FOUND")
        print(f"  All {validation_report['total_cells']:,} cell barcodes are globally unique")
    
    # Final summary
    print("\n" + "=" * 70)
    print("VALIDATION SUMMARY")
    print("=" * 70)
    print(f"\nTotal cells: {validation_report['total_cells']:,}")
    print(f"Unique barcodes: {validation_report['unique_barcodes']:,}")
    
    if validation_report['duplicates_found']:
        print(f"\nVALIDATION FAILED")
        print(f"  {len(duplicate_barcodes):,} duplicate barcodes found")
        print(f"  {validation_report['total_cells'] - validation_report['unique_barcodes']:,} cells affected")
        print(f"\nRECOMMENDED ACTIONS:")
        print(f"  1. Add unique dataset prefix to all barcodes")
        print(f"  2. Use: adata.obs_names = 'DATASET_' + adata.obs_names")
        print(f"  3. Re-validate before concatenation")
    else:
        print(f"\nVALIDATION PASSED")
        print(f"All cell barcodes are globally unique")
        print(f"Safe to proceed with dataset concatenation")
    
    return validation_report


validation_report = validate_cell_barcodes_global_uniqueness(adatas_dict, verbose=True)
print("\nFINAL VALIDATION REPORT:")
print(validation_report)
```

# Construct Patient Ids
* Patient id represents true patients in all datasets except for GSE151530 where the 37 id's haven't been provided by the authors, the 46 samples ids have been used instead.

```python
def create_standard_donor_id_column(
    adatas_dict: Dict,
    dataset_key: str,
    patient_column: str,
    verbose: bool = True
) -> Dict:
   
    if verbose:
        print(f"\n Creating donor_id for {dataset_key}")

    adata = None
    actual_key = None
    adata = adatas_dict[dataset_key]
    actual_key = dataset_key
    
    # Check if patient column exists (case-insensitive)
    available_columns = adata.obs.columns.tolist()
    patient_col_found = None
    
    for col in available_columns:
        if col.lower() == patient_column.lower():
            patient_col_found = col
            break
    
    if patient_col_found is None:
        print(f"'{patient_column}' column not found")
        print(f"  Available columns: {available_columns}")
        
        return adatas_dict
    
    # Create donor_id column
    adata.obs['donor_id'] = adata.obs[patient_col_found].astype(str)
    
    # Statistics
    n_donors = adata.obs['donor_id'].nunique()
    sample_ids = adata.obs['donor_id'].unique()[:5].tolist()
    
    if verbose:
        print(f"\n  Successfully created 'donor_id' column")
        print(f"    • Source column: '{patient_col_found}'")
        print(f"    • Unique donors: {n_donors}")
        print(f"    • Sample donor IDs: {sample_ids}")
        
    # Update dictionary
    adatas_dict[actual_key] = adata
```

```python
# SET PATIENT COLUMN AS DONOR_ID FOR GSE149614
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_149614',
    patient_column='patient',
    verbose=True)

# SET PATIENT COLUMN AS DONOR_ID FOR GSE151530
# Setting sample id (46 samples) as donor_id since the metadata doesn't have patient ids for the 37 patient 
"""
   True patient identifiers unavailable in GEO; "
   each sample treated as an independent donor.
   """
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_151530',
    patient_column='S_ID',
    verbose=True)
# SET PATIENT COLUMN AS DONOR_ID FOR GSE124395

create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_124395',
    patient_column='patient_id',
    verbose=True)

# SET PATIENT COLUMN AS DONOR_ID FOR GSE125449
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_125449',
    patient_column='Sample',
    verbose=True)

# SKIPPING GSE124395 AS IT DOESN'T HAVE METADATA

# SET PATIENT COLUMN AS DONOR_ID FOR GSE115469
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_115469',
    patient_column='Sample',
    verbose=True)

# SET PATIENT COLUMN AS DONOR_ID FOR GSE140228
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_140228',
    patient_column='Donor',
    verbose=True)

# SET PATIENT COLUMN AS DONOR_ID FOR GSE146409
create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_146409',
    patient_column='human',
    verbose=True)
# SET condition+sample_num COLUMN AS DONOR_ID FOR GSE136103

create_standard_donor_id_column(
    adatas_dict,
    dataset_key='adata_136103',
    patient_column='patient_id',
    verbose=True)

# SET PATIENT COLUMN AS DONOR_ID FOR GSE138709
create_standard_donor_id_column(
    adatas_dict,        
    dataset_key='adata_138709',
    patient_column='patient_id',    
    verbose=True)
```

```python
def investigate_patient_donor_ids(adatas_dict: Dict) -> pd.DataFrame:
    investigation_results = []
    
    for name, adata in adatas_dict.items():
        print(f"\n📋 Dataset: {name}")
        print("-" * 60)
        print(f"Shape: {adata.n_obs:,} cells × {adata.n_vars:,} genes")
        
        # Display ALL metadata columns
        print(f"\n📊 ALL METADATA COLUMNS ({len(adata.obs.columns)}):")
        for col in adata.obs.columns:
            n_unique = adata.obs[col].nunique()
            print(f"  • {col}: {n_unique} unique values")
        
        # Look for potential patient/donor ID columns
        potential_id_cols = []
        patient_keywords = ['patient', 'donor', 'sample', 'subject', 'individual', 'id']
        
        for col in adata.obs.columns:
            col_lower = col.lower()
            if any(keyword in col_lower for keyword in patient_keywords):
                potential_id_cols.append(col)  
        
        # Store results
        investigation_results.append({
            'dataset': name,
            'n_cells': adata.n_obs,
            'n_metadata_cols': len(adata.obs.columns),
            'potential_id_columns': ', '.join(potential_id_cols) if potential_id_cols else 'None',
            'n_potential_ids': len(potential_id_cols)
        })
        
        print()
    
    # Create summary DataFrame
    summary_df = pd.DataFrame(investigation_results)
    
    print("\n" + "=" * 80)
    print("📊 SUMMARY TABLE")
    print("=" * 80)
    print(summary_df.to_string(index=False))
    
    return summary_df

# Execute investigation
print("🔍 STEP 1: INVESTIGATE ALL DATASETS")
summary = investigate_patient_donor_ids(adatas_dict)
```

# Add unified metadata columms; donor resolution, compartment, tissue, tumor_status, disease_status, cancer_type, disease_group


```python
def add_donor_resolution_metadata(adatas_dict: Dict, verbose: bool = True) -> Dict:
    """
    Add 'donor_resolution' column to all datasets indicating the source/quality 
    of donor ID information.
    Categories:
    - 'patient': True biological patient IDs from metadata
    - 'sample': Sample IDs used as proxy (no patient info available)
    - 'constructed': Donor ID built from multiple metadata columns
    - 'unknown': No donor information available
    """
    
    # Define donor resolution for each dataset based on your documentation
    donor_resolution_map = {
        # Datasets with true patient IDs
        'adata_149614': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        },
        'adata_125449': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        }, 
        'adata_115469': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        },
        'adata_140228': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        },
        'adata_146409': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        },
        'adata_138709': {
            'resolution': 'patient',
            'description': 'True patient identifiers'
        },
        
        # Datasets using sample IDs as proxy
        'adata_151530': {
            'resolution': 'sample',
            'description': 'Sample IDs used (46 samples, 37 patients unavailable)'
        },
        
        # Datasets with constructed donor IDs
        'adata_136103': {
            'resolution': 'constructed',
            'description': 'Reconstructed from disease condition and sample number'
        },
        
        # Datasets with no donor information
        'adata_124395': {
            'resolution': 'unknown',
            'source_column': 'N/A',
            'description': 'No patient/donor metadata available'
        }
    }
    
    # Process each dataset
    summary_data = []
    
    for name, adata in adatas_dict.items():
        # Get resolution info for this dataset
        if name in donor_resolution_map:
            resolution_info = donor_resolution_map[name]
            resolution = resolution_info['resolution']
            description = resolution_info['description']
        else: 
            print("Unknown dataset")
        
        # Add donor_resolution column
        adata.obs['donor_resolution'] = resolution
        
        # Add detailed metadata to uns
        adata.uns['donor_metadata'] = {
            'resolution': resolution,
            'description': description
        }
        
    
        # Collect summary statistics
        summary_data.append({
            'Dataset': name,
            'Resolution': resolution,
            'N_Cells': adata.n_obs
        })
        
        # Update dictionary
        adatas_dict[name] = adata
    
    # Print summary table
    if verbose:
        print("\n" + "=" * 70)
        print("📊 DONOR RESOLUTION SUMMARY")
        print("=" * 70)
        
        summary_df = pd.DataFrame(summary_data)
        print(summary_df.to_string(index=False))
        
        # Category breakdown
        print("\n📈 RESOLUTION CATEGORIES:")
        resolution_counts = summary_df['Resolution'].value_counts()
        for resolution, count in resolution_counts.items():
            print(f"  • {resolution}: {count} dataset(s)")
        
    return adatas_dict
adatas_dict = add_donor_resolution_metadata(adatas_dict, verbose=True)
```

```python
import pandas as pd

def annotate_gse146409(adata):
    """
    liver tumor microenvironment atlas (MARS-seq).
    Key fields used:
      - obs['treatment'] : {'tumor','far'}  where 'far' = non-malignant liver tissue far from tumor
      - obs['plate']     : strings like 'tumor1', 'far3', etc. (technical plate/library)
     """ 

    # Preconditions + cleaning
    required = ["treatment", "plate"]
    missing = [c for c in required if c not in adata.obs.columns]
    if missing:
        raise KeyError(f"columns {missing}, are missing.")

    tr = adata.obs["treatment"].astype(str).str.strip().str.lower()
    pl = adata.obs["plate"].astype(str).str.strip().str.lower()

    # Tissue + high-level disease semantics
    # All samples in this series are liver resections (malignant and matched non-malignant liver).
    adata.obs["tissue"] = "liver"

    # Disease-level labels:
    # Patients underwent liver resection for malignancy/lesion; "far" is non-tumor tissue but from a disease context.
    # Use "diseased" because donors are not healthy controls.
    adata.obs["disease_status"] = "diseased"

    # Cancer type at dataset-level is mixed (cholangiocarcinoma + colorectal liver metastasis + benign cyst per GEO design).
    # But the per-donor pathology labels weren't provided.
    adata.obs["cancer_type"] = "mixed"         # more conservative to prevent guessing
    adata.obs["disease_group"] = "mixed"       # controlled vocabulary for mixed liver lesions

    
    # Tumor vs non-tumor (far)
    # 'far' is non-malignant liver tissue far from tumor.
    adata.obs["tumor_status"] = "unknown"
    adata.obs.loc[tr.eq("tumor"), "tumor_status"] = "tumor"
    adata.obs.loc[tr.eq("far"),   "tumor_status"] = "non_tumor"

    # Compartment
    # I'm keeping the anatomical compartment simple: tumor vs far(non-tumor).
    adata.obs["compartment"] = "unknown"
    adata.obs.loc[tr.eq("tumor"), "compartment"] = "tumor"
    adata.obs.loc[tr.eq("far"),   "compartment"] = "adjacent" # to maintain unified terms

    
    # plate examples: tumor1..tumor10, far1..far15.
    # We'll create:
    #   plate_id: original plate value
    #   plate_group: 'tumor'/'far'/unknown inferred from plate prefix 
    #   sample_id: plate_id 
    adata.obs["plate_id"] = pl

    plate_group = pd.Series("unknown", index=adata.obs_names, dtype="object")
    plate_group.loc[pl.str.startswith("tumor")] = "tumor"
    plate_group.loc[pl.str.startswith("far")]   = "far"
    adata.obs["plate_group"] = plate_group.values

    # sample_id for integration: treat each plate/library as sample (technical unit)
    # (This dataset is plate-based; plate is the cleanest library identifier.)
    adata.obs["sample_id"] = adata.obs["plate_id"]

   
    print("treatment:\n", adata.obs["treatment"].astype(str).str.lower().value_counts(dropna=False))
    print("plate_group:\n", adata.obs["plate_group"].value_counts(dropna=False))
    print("compartment:\n", adata.obs["compartment"].value_counts(dropna=False))
    print("tissue:\n", adata.obs["tissue"].value_counts(dropna=False))
    print("tumor_status:\n", adata.obs["tumor_status"].value_counts(dropna=False))
    print("donor_id (nunique):", adata.obs["donor_id"].nunique(dropna=False))
    print("sample_id (nunique):", adata.obs["sample_id"].nunique(dropna=False))

    return adata

adatas_dict["adata_146409"] = annotate_gse146409(adatas_dict["adata_146409"])
```

# HARMONIZE CELL TYPE ANNOTATIONS

A systematic cell-type harmonization was performed across all single-cell RNA-seq datasets included in the liver atlas. For datasets providing study-specific cell-type annotations, original labels were first standardized into a unified raw annotation field (cell_type_raw), preserving the original study nomenclature without modification. We then mapped these labels into a controlled, hierarchical cell-type ontology comprising broad compartments (Level 0: immune, epithelial, endothelial, stromal, and other) and biologically interpretable canonical cell types (Level 1), such as T cells, B cells, myeloid cells, hepatocytes, cholangiocytes, endothelial cells, fibroblast/CAF populations, stellate cells, and malignant epithelial cells. Where available and biologically meaningful, finer-grained annotations were retained as an optional Level-2 label to support within-lineage analyses while maintaining comparability at higher levels.

Harmonization decisions were intentionally conservative and prioritized biological correctness over maximal granularity. Cell-type labels that were inherently ambiguous or dataset-specific (e.g., progenitor-like or specialized tumor microenvironment populations) were mapped to the most appropriate parent class at Level 1, while preserving the original label for transparency and potential refinement. No cell-type labels were inferred in datasets lacking explicit annotations, and no forced reassignment was applied to unresolved or low-confidence populations, which were retained as unclassified. Proliferative states were identified using label-based indicators and flagged separately rather than redefining lineage identity. This approach ensured that harmonized labels were consistent, reproducible, and suitable for downstream integration, scIB benchmarking, and trajectory inference analyses without introducing unsupported biological assumptions.

A small number of dataset-specific cell-type labels could not be unambiguously resolved at a fine-grained lineage level based solely on study-provided annotations. In these cases, we adopted a deliberately conservative harmonization strategy by assigning such populations to the most appropriate canonical parent class (Level 1) while retaining the original study label as a finer-resolution annotation (Level 2). For example, in the liver tumor microenvironment dataset GSE146409, the labels “SAMs” and “TM1”, which likely represent specialized macrophage or tumor-associated myeloid programs, were mapped to the myeloid compartment at the canonical level, while preserving their original labels for transparency and potential refinement. Similarly, vascular populations annotated as “LVEC” or “LVECt” were assigned to the endothelial compartment at Level 1, without asserting more specific vascular subtypes in the absence of marker-based confirmation. In the tumor biodiversity dataset GSE125449, cells annotated as “HPC-like” were conservatively mapped to malignant epithelial cells, reflecting a progenitor-like epithelial tumor state rather than a definitive normal hepatocyte or cholangiocyte identity, which would require additional molecular evidence. Importantly, these assignments represent broad compartment-level mappings intended for integration and benchmarking, and do not imply definitive subtype identity. By preserving original labels at a finer level, this approach avoids over-interpretation while maintaining flexibility for future marker-based validation and subtype-specific analyses.

## Diagnose normalized dataset and apply normalization and logp for raw counts  

```python

# - Detects scale on BOTH:
#     (1) a trusted counts layer if available (default: layers['counts'])
#     (2) adata.X (the active matrix)
# - Never normalizes log/processed matrices.
# - If state is uncertain:
#     - normalize ONLY if a trusted counts layer exists
#     - otherwise SKIP and warn 
#


from __future__ import annotations

import io
import sys
import warnings
import logging
from dataclasses import dataclass
from typing import Optional, Dict, Any

import numpy as np
import scipy.sparse as sp
import scanpy as sc

import pandas as pd 

def _display_norm_table(rows):
    df = pd.DataFrame(rows)

    cols = [
        "dataset", "action", "X_status",
        "counts_layer_present", "counts_status",
        "X_min", "X_max", "X_frac_frac", "X_frac_(0,1)",
        "counts_min", "counts_max", "counts_frac_frac",
        "reason"
    ]
    df = df[cols]

    # Pretty printing for both notebooks + terminals
    with pd.option_context("display.max_rows", None, "display.float_format", "{:.3g}".format):
        try:
            display(df)  # Jupyter
        except NameError:
            pass
        print(df.to_string(index=False))

def _sample_values_from_matrix(M, n_sample: int = 200_000, seed: int = 0) -> np.ndarray:
    """Return a 1D float array of sampled values from dense or sparse matrix M."""
    rng = np.random.default_rng(seed)

    if sp.issparse(M):
        data = M.data
        if data.size == 0:
            return np.zeros(1000, dtype=float)
        k = min(n_sample, data.size)
        return rng.choice(data, size=k, replace=False).astype(float)
    else:
        arr = np.asarray(M).ravel()
        if arr.size == 0:
            return np.zeros(1000, dtype=float)
        k = min(n_sample, arr.size)
        return rng.choice(arr, size=k, replace=False).astype(float)

def _dense_block_for_checks(M, n_cells: int, n_genes: int) -> np.ndarray:
    """Small dense block for integer/fraction checks."""
    if sp.issparse(M):
        return M[:n_cells, :n_genes].toarray().astype(float)
    return np.asarray(M[:n_cells, :n_genes]).astype(float)

@dataclass
class ScaleDiagnostics:
    layer: str
    n_vals_sampled: int
    min_val: float
    max_val: float
    frac_neg: float
    frac_zero: float
    frac_integer_like_nonzero: float
    frac_fractional_nonzero: float
    frac_between_0_1_nonzero: float
    note: str


def detect_matrix_scale(
    adata,
    layer: Optional[str] = None,
    sample_size: int = 200_000,
    seed: int = 0,
    block_cells: int = 100,
    block_genes: int = 200,
    int_atol: float = 1e-6,
) -> tuple[str, ScaleDiagnostics]:
    """
    Detect if a matrix looks like:
      - 'raw_counts'           (non-negative, mostly integer-like, and/or large range)
      - 'already_normalized'   (many fractional nonzeros, compressed range, etc.)
      - 'uncertain'

    Returns: (status, diagnostics)
    """
    M = adata.layers[layer] if layer is not None else adata.X
    layer_name = f"layers['{layer}']" if layer is not None else "X"

    vals = _sample_values_from_matrix(M, n_sample=sample_size, seed=seed)
    if vals.size == 0:
        diag = ScaleDiagnostics(
            layer=layer_name, n_vals_sampled=0, min_val=np.nan, max_val=np.nan,
            frac_neg=np.nan, frac_zero=np.nan,
            frac_integer_like_nonzero=np.nan, frac_fractional_nonzero=np.nan,
            frac_between_0_1_nonzero=np.nan,
            note="Empty matrix sample."
        )
        return "uncertain", diag

    min_val = float(np.nanmin(vals))
    max_val = float(np.nanmax(vals))

    frac_neg = float(np.mean(vals < 0))
    frac_zero = float(np.mean(vals == 0))

    # Use a dense block for nonzero/fraction checks (more stable than sparse .data alone)
    bc = min(block_cells, adata.n_obs)
    bg = min(block_genes, adata.n_vars)
    block = _dense_block_for_checks(M, bc, bg)

    if np.any(~np.isfinite(block)):
        warnings.warn(f"[{layer_name}] NaN/Inf detected in expression matrix. Fix before preprocessing.")

    nz = block[block > 0]
    if nz.size < 20:
        # too sparse in the small block to be sure
        frac_int_like = np.nan
        frac_frac = np.nan
        frac_0_1 = np.nan
    else:
        frac_int_like = float(np.mean(np.isclose(nz, np.round(nz), atol=int_atol)))
        frac_frac = float(np.mean(~np.isclose(nz, np.round(nz), atol=int_atol)))
        frac_0_1 = float(np.mean((nz > 0) & (nz < 1)))

    # Strong signals for processed/log/normalized:
    # - negative values (z-score / scaled / centered)
    # - substantial fractional nonzeros AND compressed max
    if min_val < 0: 
        status = "already_normalized"
        note = "Negative values detected (scaled / centered / log/processed)."
    else:
        # candidate processed: many fractional nonzeros + compressed range
        processed_like = (
            (not np.isnan(frac_frac)) and (frac_frac >= 0.20) and (max_val <= 20.0)
        )

        # candidate raw: non-negative, mostly integer-like nonzeros, and either large max or very low fractional rate
        raw_like = (
            (not np.isnan(frac_int_like)) and (frac_int_like >= 0.90) and (
                (max_val >= 50.0) or ((not np.isnan(frac_frac)) and (frac_frac <= 0.05))
            )
        )

        if processed_like and not raw_like:
            status = "already_normalized"
            note = "Fractional nonzeros + compressed range (<=20) suggest log/normalized."
        elif raw_like and not processed_like:
            status = "raw_counts"
            note = "Integer-like nonzeros + large dynamic range suggest raw counts."
        else:
            status = "uncertain"
            note = "Signals mixed or insufficient; classification uncertain."

    diag = ScaleDiagnostics(
        layer=layer_name,
        n_vals_sampled=int(vals.size),
        min_val=min_val,
        max_val=max_val,
        frac_neg=frac_neg,
        frac_zero=frac_zero,
        frac_integer_like_nonzero=float(frac_int_like) if not np.isnan(frac_int_like) else np.nan,
        frac_fractional_nonzero=float(frac_frac) if not np.isnan(frac_frac) else np.nan,
        frac_between_0_1_nonzero=float(frac_0_1) if not np.isnan(frac_0_1) else np.nan,
        note=note,
    )
    return status, diag


def _suppress_scanpy_output():
    """Context manager-ish setup to suppress scanpy stdout + logger spam."""
    sc_logger = logging.getLogger("scanpy")
    ad_logger = logging.getLogger("anndata")
    original_levels = (sc_logger.level, ad_logger.level)
    original_verbosity = sc.settings.verbosity
    original_stdout = sys.stdout

    sc_logger.setLevel(logging.CRITICAL)
    ad_logger.setLevel(logging.CRITICAL)
    sc.settings.verbosity = 0
    sys.stdout = io.StringIO()
    return original_levels, original_verbosity, original_stdout


def _restore_scanpy_output(original_levels, original_verbosity, original_stdout):
    sc_logger = logging.getLogger("scanpy")
    ad_logger = logging.getLogger("anndata")
    sc_logger.setLevel(original_levels[0])
    ad_logger.setLevel(original_levels[1])
    sc.settings.verbosity = original_verbosity
    sys.stdout = original_stdout


def normalize_and_log1p(
    adata,
    target_sum: float = 1e4,
    counts_layer: Optional[str] = None,
    save_raw_to_layer: Optional[str] = "counts",
) -> None:
    """
    Normalize total and log1p on adata.X.
    - If counts_layer is given, sets X from that layer BEFORE normalization.
    - If save_raw_to_layer is given, stores raw counts snapshot BEFORE normalization.
    """
    # Optionally set X from trusted counts layer
    if counts_layer is not None:
        if counts_layer not in adata.layers:
            raise KeyError(f"counts_layer='{counts_layer}' not found in adata.layers.")
        adata.X = adata.layers[counts_layer].copy()

    # Save raw snapshot
    if save_raw_to_layer is not None:
        if save_raw_to_layer not in adata.layers:
            adata.layers[save_raw_to_layer] = adata.X.copy()

    # Ensure non-negative
    if sp.issparse(adata.X):
        minv = adata.X.min()
    else:
        minv = float(np.nanmin(adata.X))
    if minv < 0:
        raise ValueError("Negative values in X; refusing to normalize/log. Data is already processed or corrupted.")

    # Apply normalize_total + log1p (quiet)
    original_levels, original_verbosity, original_stdout = _suppress_scanpy_output()
    try:
        sc.pp.normalize_total(adata, target_sum=target_sum, inplace=True)
        sc.pp.log1p(adata)
    finally:
        _restore_scanpy_output(original_levels, original_verbosity, original_stdout)


def safe_normalize_log_pipeline(
    adata,
    dataset_key: Optional[str] = None,
    target_sum: float = 1e4,
    trusted_counts_layer: str = "counts",
    allow_normalize_if_uncertain_without_counts: bool = False,
    write_provenance: bool = True,
    verbose: bool = True,
) -> Dict[str, Any]:
    """
    Decide whether to normalize/log1p based on:
      - adata.layers[trusted_counts_layer] if exists (trusted counts)
      - adata.X scale (active matrix)
    Policy:
      - If X already looks normalized/log: SKIP (even if counts exist)
      - If X looks raw:
          - If counts layer exists, normalize from counts (best)
          - Else normalize in-place from X
      - If uncertain:
          - If counts layer exists: normalize from counts (safe-ish)
          - Else SKIP unless allow_normalize_if_uncertain_without_counts=True

    Returns a provenance dict (and optionally writes it to adata.uns).
    """
    if dataset_key is None:
        dataset_key = str(adata.obs.get("dataset", "unknown").iloc[0]) if "dataset" in adata.obs else "unknown"

    # Detect on X
    x_status, x_diag = detect_matrix_scale(adata, layer=None)

    # Detect on trusted counts layer if present
    has_counts_layer = trusted_counts_layer in adata.layers
    c_status, c_diag = (None, None)
    if has_counts_layer:
        c_status, c_diag = detect_matrix_scale(adata, layer=trusted_counts_layer)

    action = "skipped"
    reason = ""

    if x_status == "already_normalized":
        action = "skipped"
        reason = "X appears already normalized/log/processed; skipping to avoid double-transform."
    elif x_status == "raw_counts":
        # Prefer normalizing from trusted counts layer if it exists and is raw
        if has_counts_layer and c_status == "raw_counts":
            normalize_and_log1p(
                adata,
                target_sum=target_sum,
                counts_layer=trusted_counts_layer,
                save_raw_to_layer=trusted_counts_layer,  # keep counts snapshot
            )
            action = "normalized_from_counts_layer"
            reason = f"X looks raw; used trusted counts layer '{trusted_counts_layer}'."
        else:
            # Normalize from X directly
            normalize_and_log1p(
                adata,
                target_sum=target_sum,
                counts_layer=None,
                save_raw_to_layer=(trusted_counts_layer if not has_counts_layer else None),
            )
            action = "normalized_from_X"
            reason = "X looks raw; normalized/log1p applied directly to X."
    else:
        # x_status == uncertain
        if has_counts_layer and c_status == "raw_counts":
            normalize_and_log1p(
                adata,
                target_sum=target_sum,
                counts_layer=trusted_counts_layer,
                save_raw_to_layer=trusted_counts_layer,
            )
            action = "normalized_from_counts_layer"
            reason = f"X uncertain; but trusted counts layer '{trusted_counts_layer}' looks raw."
        else:
            if allow_normalize_if_uncertain_without_counts:
                normalize_and_log1p(
                    adata,
                    target_sum=target_sum,
                    counts_layer=None,
                    save_raw_to_layer=(trusted_counts_layer if not has_counts_layer else None),
                )
                action = "normalized_from_X"
                reason = "X uncertain; forced normalization/log1p per user policy."
            else:
                action = "skipped_uncertain"
                reason = "X uncertain and no verified counts layer; skipping."

    prov = {
        "dataset_key": dataset_key,
        "action": action,
        "reason": reason,
        "target_sum": float(target_sum),
        "trusted_counts_layer": trusted_counts_layer,
        "X_detection": x_status,
        "X_diag": x_diag.__dict__,
        "counts_layer_present": bool(has_counts_layer),
        "counts_detection": c_status,
        "counts_diag": (c_diag.__dict__ if c_diag is not None else None),
    }

    
    if verbose:
        print(f"[{dataset_key}] action={action} | X={x_status} | counts_layer={has_counts_layer} ({c_status})")
        print(f"  reason: {reason}")
        print(f"  X: min={x_diag.min_val:.3g}, max={x_diag.max_val:.3g}, "
              f"frac_frac_nonzero={x_diag.frac_fractional_nonzero:.3f}, frac_0_1_nonzero={x_diag.frac_between_0_1_nonzero:.3f}")
        if has_counts_layer and c_diag is not None:
            print(f"  counts: min={c_diag.min_val:.3g}, max={c_diag.max_val:.3g}, "
                  f"frac_frac_nonzero={c_diag.frac_fractional_nonzero:.3f}, frac_0_1_nonzero={c_diag.frac_between_0_1_nonzero:.3f}")

    return prov
def run_pipeline_for_adatas_dict(
    adatas_dict: Dict[str, Any],
    keys: Optional[list[str]] = None,
    target_sum: float = 1e4,
    trusted_counts_layer: str = "counts",
    allow_normalize_if_uncertain_without_counts: bool = False,
):
    if keys is None:
        keys = list(adatas_dict.keys())

    rows = []

    for k in keys:
        a = adatas_dict[k]

        # IMPORTANT: verbose=False prevents the redundant print blocks
        prov = safe_normalize_log_pipeline(
            a,
            dataset_key=k,
            target_sum=target_sum,
            trusted_counts_layer=trusted_counts_layer,
            allow_normalize_if_uncertain_without_counts=allow_normalize_if_uncertain_without_counts,
            write_provenance=False,
            verbose=False,
        )

        x = prov["X_diag"]
        c = prov["counts_diag"] if prov["counts_diag"] is not None else {}

        rows.append({
            "dataset": k,
            "action": prov["action"],
            "X_status": prov["X_detection"],
            "counts_layer_present": prov["counts_layer_present"],
            "counts_status": prov["counts_detection"],
            "X_min": x["min_val"],
            "X_max": x["max_val"],
            "X_frac_frac": x["frac_fractional_nonzero"],
            "X_frac_(0,1)": x["frac_between_0_1_nonzero"],
            "counts_min": c.get("min_val", None),
            "counts_max": c.get("max_val", None),
            "counts_frac_frac": c.get("frac_fractional_nonzero", None),
            "reason": prov["reason"],
        })

    _display_norm_table(rows)
    
run_pipeline_for_adatas_dict(
    adatas_dict,
    keys=list(adatas_dict.keys()),
    target_sum=1e4,
    trusted_counts_layer="counts",
    allow_normalize_if_uncertain_without_counts=False,
)
```

```python
def verify_counts_matrix(M, n=200_000, seed=0):
    rng = np.random.default_rng(seed)

    # sample values
    if sp.issparse(M):
        data = M.data
        if data.size == 0:
            return {"ok": False, "reason": "all zeros"}
        k = min(n, data.size)
        vals = rng.choice(data, size=k, replace=False).astype(float)
    else:
        arr = np.asarray(M).ravel()
        k = min(n, arr.size)
        vals = rng.choice(arr, size=k, replace=False).astype(float)

    # diagnostics
    frac_int = np.mean(np.isclose(vals, np.round(vals), atol=1e-6))
    frac_neg = np.mean(vals < 0)
    frac_between_0_1 = np.mean((vals > 0) & (vals < 1))
    mx = float(np.max(vals))
    mn = float(np.min(vals))

    ok = (frac_neg == 0) and (frac_int > 0.98) and (frac_between_0_1 < 0.01)
    reason = f"frac_int={frac_int:.3f}, frac_between_0_1={frac_between_0_1:.3f}, min={mn:.3g}, max={mx:.3g}"

    return {"ok": ok, "reason": reason, "frac_int": frac_int, "frac_between_0_1": frac_between_0_1, "min": mn, "max": mx}

def verify_adata_counts(adata, name):
    print(f"\n=== {name} ===")
    print("X:", verify_counts_matrix(adata.X))
    if "counts" in adata.layers:
        print("counts layer:", verify_counts_matrix(adata.layers["counts"]))
verify_adata_counts(adatas_dict["adata_124395"], "adata_124395")
```

```python
import numpy as np
import scipy.sparse as sp
def fractional_diagnostics(adata, layer=None, n_cells=500, n_genes=500, tol=1e-6):
    X = adata.layers[layer] if layer is not None else adata.X
    Xs = X[:min(n_cells, adata.n_obs), :min(n_genes, adata.n_vars)]

    if sp.issparse(Xs):
        v = Xs.data.astype(float, copy=False)
    else:
        v = np.asarray(Xs, dtype=float).ravel()
        v = v[v != 0]

    if v.size == 0:
        return {"nonzero": 0}

    frac_part = np.abs(v - np.round(v))
    frac_nonzero = float(np.mean(frac_part > tol))

    return {
        "nonzero": int(v.size),
        "min_value": float(v.min()),
        "max_value": float(v.max()),
        "frac_negative": float(np.mean(v < 0)),
        "frac_(0,1)": float(np.mean((v > 0) & (v < 1))),
        "frac_nonzero_fractional": frac_nonzero,
        "p99_fractional_part": float(np.quantile(frac_part, 0.99)),
        "max_fractional_part": float(frac_part.max()),
        "median_value": float(np.median(v)),
    }


print("GSE124395 X:", fractional_diagnostics(adatas_dict["adata_124395"], layer=None))
print("GSE124395 counts:", fractional_diagnostics(adatas_dict["adata_124395"], layer="counts"))
```

## Checkpointing Datasets After Normalization and Log Transformation



### Purpose
After completing quality control, cell-type harmonization, and normalization/log-transformation, we create a **checkpoint of all datasets** before proceeding to memory-intensive steps such as gene intersection, concatenation, and highly variable gene (HVG) selection. This checkpoint allows the analysis to be resumed from a stable intermediate state without re-running all upstream preprocessing, and prevents kernel crashes caused by large intermediate objects.

---

### Why a Checkpoint Is Necessary
Atlas-level single-cell RNA-seq analyses involve repeated subsetting and concatenation operations.

Without careful memory management, these operations can exceed available RAM and crash the computational environment. Saving datasets at this stage provides a **safe recovery point** and enables iterative experimentation with downstream integration strategies.

---

### Why Expression Matrices Are Stored in CSR Sparse Format
Single-cell expression matrices are **highly sparse**, typically containing 90–99% zero values. Storing such matrices in dense format is inefficient and can require tens of gigabytes of memory for large atlases.

We therefore store expression matrices in **Compressed Sparse Row (CSR)** format:

- CSR stores only non-zero values and their indices.
- Memory usage is reduced by orders of magnitude.
- CSR supports efficient row and column slicing.


### Handling of Raw Counts and Normalized Expression
- The main expression matrix (`adata.X`) contains **normalized and log-transformed expression** used for integration, visualization, and trajectory inference.
- When available, **raw count data are preserved separately** in `adata.layers["counts"]` for methods that require counts (e.g. scVI/scANVI).
- Both `X` and `layers["counts"]` are stored in CSR format to ensure memory efficiency.

```python
# CHECKPOINT: Save normalized/log1p datasets before intersection/HVG
#   - stores per-dataset .h5ad
#   - enforces sparse CSR to reduce memory
#   - float32 downcast
#   - writes a manifest for quick auditing


def _to_csr(X):
    if sp.issparse(X):
        return X.tocsr()
    return sp.csr_matrix(X)

def _maybe_float32(X):
    if sp.issparse(X):
        if np.issubdtype(X.dtype, np.floating) and X.dtype != np.float32:
            return X.astype(np.float32)
        return X
    else:
        if np.issubdtype(X.dtype, np.floating) and X.dtype != np.float32:
            return X.astype(np.float32)
        return X

def _fix_obs_for_h5ad(adata: ad.AnnData) -> None:
    """
    Make obs safely writable:
      - convert pandas nullable strings to object
      - ensure categorical categories are plain object
    """
    for col in adata.obs.columns:
        s = adata.obs[col]

        # Nullable string columns -> object
        if pd.api.types.is_string_dtype(s.dtype) and not pd.api.types.is_object_dtype(s.dtype):
            adata.obs[col] = s.astype("object")
            s = adata.obs[col]

        # Categorical categories -> object
        if pd.api.types.is_categorical_dtype(s.dtype):
            cats = s.cat.categories
            if pd.api.types.is_string_dtype(cats.dtype) and not pd.api.types.is_object_dtype(cats.dtype):
                adata.obs[col] = s.cat.set_categories(cats.astype("object"))

def _sanitize_h5_keys(name: str) -> str:
    """HDF5 forbids '/' in keys; replace with '_' (and also backslash for safety)."""
    return name.replace("/", "_").replace("\\", "_")

def _sanitize_obs_columns_for_h5ad(adata: ad.AnnData) -> dict:
    """
    Rename obs columns that contain forbidden characters for HDF5 storage (e.g., '/').
    Returns mapping {old: new} for changed columns.
    """
    mapping = {}
    new_cols = []
    seen = set()

    for c in adata.obs.columns:
        nc = _sanitize_h5_keys(str(c))
        # Avoid collisions if two columns map to same sanitized name
        base = nc
        i = 1
        while nc in seen:
            i += 1
            nc = f"{base}__{i}"
        seen.add(nc)
        new_cols.append(nc)
        if nc != c:
            mapping[str(c)] = nc

    if mapping:
        adata.obs.columns = new_cols
    return mapping

def checkpoint_save_adatas_dict(
    adatas_dict: dict,
    out_dir: str | Path,
    keys: list[str] | None = None,
    make_X_csr: bool = True,
    make_counts_csr: bool = True,
    downcast_float32: bool = True,
    compression: str = "gzip",
):
    out_dir = Path(out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    if keys is None:
        keys = sorted(list(adatas_dict.keys()))

    # Opt-in to writing nullable strings 
    ad.settings.allow_write_nullable_strings = True

    manifest_rows = []
    for k in keys:
        a = adatas_dict[k].copy()
        fname = out_dir / f"{k}.h5ad"

        # Ensure dataset key is stored
        a.uns.setdefault("checkpoint", {})
        a.uns["checkpoint"]["dataset_key"] = k
        a.uns["checkpoint"]["checkpoint_stage"] = "post_norm_log1p_pre_intersection"

        # X handling
        if make_X_csr:
            a.X = _to_csr(a.X)
        if downcast_float32:
            a.X = _maybe_float32(a.X)

        # counts layer handling (if exists)
        counts_exists = "counts" in a.layers
        if counts_exists:
            if make_counts_csr:
                a.layers["counts"] = _to_csr(a.layers["counts"])
            if downcast_float32:
                a.layers["counts"] = _maybe_float32(a.layers["counts"])

        # Fix obs dtypes so h5ad can write them
        _fix_obs_for_h5ad(a)

        # Sanitize obs column names (removes '/' etc.)
        renamed = _sanitize_obs_columns_for_h5ad(a)
        if renamed:
            # We can't store dict keys with '/' inside uns (HDF5 forbids it)
            a.uns["checkpoint"]["renamed_obs_columns_table"] = pd.DataFrame(
                {"old": list(renamed.keys()), "new": list(renamed.values())}
            )

        # Write
        a.write_h5ad(fname, compression=compression)

        manifest_rows.append({
            "dataset_key": k,
            "file": fname.name,
            "n_obs": a.n_obs,
            "n_vars": a.n_vars,
            "X_sparse": sp.issparse(a.X),
            "X_dtype": str(a.X.dtype),
            "counts_layer": counts_exists,
            "counts_sparse": (sp.issparse(a.layers["counts"]) if counts_exists else False),
            "counts_dtype": (str(a.layers["counts"].dtype) if counts_exists else ""),
            "renamed_obs_cols": len(renamed),
        })

        msg = f"[SAVED] {k}: {a.n_obs:,} × {a.n_vars:,} -> {fname.name}"
        if renamed:
            msg += f" | obs renamed: {len(renamed)}"
        print(msg)

    manifest = pd.DataFrame(manifest_rows).sort_values("dataset_key")
    manifest_file = out_dir / "CHECKPOINT_manifest.tsv"
    manifest.to_csv(manifest_file, sep="\t", index=False)
    print(f"\nManifest saved: {manifest_file}")
    return manifest

checkpoint_dir = "/data1/esraa/Thesis-Project/Data/Processed_data/post_norm_log1p"

# Save all 9 datasets 
manifest = checkpoint_save_adatas_dict(
    adatas_dict=adatas_dict,
    out_dir=checkpoint_dir,
    keys=[ 
        "adata_149614","adata_151530","adata_125449","adata_115469",
        "adata_140228","adata_146409","adata_124395","adata_138709","adata_136103"
    ],  
    make_X_csr=True,
    make_counts_csr=True,
    downcast_float32=True,
    compression="gzip",
)
```

# Intersect genes → concatenate → HVG selection with batch_key:



Gene selection was performed using a three-step strategy: gene intersection, dataset concatenation, and batch-aware highly variable gene (HVG) selection. First, analyses were restricted to the intersection of genes shared across all datasets included in a given benchmark. This step enforces a common feature space, which is a prerequisite for joint dimensionality reduction, batch correction, and latent-space modeling, and prevents any dataset from contributing private or platform-specific genes that could bias integration results.

Second, datasets were concatenated prior to HVG selection rather than selecting variable genes independently within each dataset. Concatenation allows gene-wise variance to be assessed in a global context while preserving dataset identity via metadata, ensuring that variability is evaluated comparably across studies. This avoids biases introduced by per-dataset HVG selection, which can overrepresent genes that are highly variable in only a subset of datasets (e.g., tumor-specific or technology-driven effects) and reduce overlap in selected features.

Finally, HVG selection was performed using a batch-aware procedure with the dataset identifier specified as the batch key. This explicitly accounts for between-dataset technical variation during variance estimation, prioritizing genes that show reproducible biological variability across datasets while down-weighting genes whose variability is primarily driven by batch effects. Together, this strategy maximizes compatibility across datasets, preserves shared biological signal, and provides a principled and reproducible feature set for downstream integration and benchmarking of batch correction and trajectory inference methods.

## Gene Intersection  


```python
from __future__ import annotations
from typing import Dict, List, Tuple, Optional
import anndata as ad
import numpy as np
import pandas as pd


def _normalize_gene_names(
    genes: pd.Index,
    upper: bool = True,
    strip: bool = True,
) -> pd.Index:
    g = genes.astype(str)
    if strip:
        g = pd.Index([x.strip() for x in g], dtype="object")
    if upper:
        g = pd.Index([x.upper() for x in g], dtype="object")
    return g


def intersect_genes_across_adatas_memory_safe(
    adatas_dict: Dict[str, ad.AnnData],
    exclude_keys: Optional[List[str]] = None,
    *,
    gene_case: str = "upper",   # "upper" or "keep"
    strip: bool = True,
    inplace: bool = True,       # True = modify objects in adatas_dict directly
    verbose: bool = True,
) -> Tuple[Dict[str, ad.AnnData], List[str]]:
    """
    Memory-safe gene intersection:
      1) computes gene intersection using only var_names (cheap)
      2) subsets each AnnData IN-PLACE (no .copy()) to avoid RAM spikes
      3) ensures identical gene order across datasets

    Returns
    -------
    (adatas_dict, common_genes)
    """ 
    exclude_keys = exclude_keys or []
    exclude_set = set(exclude_keys)
    include_keys = [k for k in adatas_dict.keys() if k not in exclude_set]

    if len(include_keys) == 0:
        raise ValueError("No datasets left after excluding keys.")

    gene_norm_mode = gene_case.lower().strip()
    use_upper = (gene_norm_mode == "upper")

    # Build per-dataset normalized gene index (deduplicated) and mapping norm->orig
    gene_sets = {}
    gene_maps = {}  # norm -> orig (first occurrence)
    gene_counts = {}

    for k in include_keys:
        a = adatas_dict[k]
        if a.var_names is None or len(a.var_names) == 0:
            raise ValueError(f"{k} has empty var_names.")

        orig = pd.Index(a.var_names.astype(str), dtype="object")
        norm = _normalize_gene_names(orig, upper=use_upper, strip=strip)

        # Deduplicate after normalization (keep first)
        if norm.has_duplicates:
            keep = ~norm.duplicated(keep="first")
            orig = orig[keep]
            norm = norm[keep]

        gene_sets[k] = set(norm.tolist())
        gene_maps[k] = pd.Series(orig.values, index=norm.values)  # norm -> orig
        gene_counts[k] = len(norm)

    # Intersection in normalized space
    common_norm = sorted(set.intersection(*[gene_sets[k] for k in include_keys]))
    if len(common_norm) == 0:
        raise ValueError(
            "Gene intersection is empty. "
            "Check gene identifiers (symbols vs Ensembl) or set gene_case='keep'."
        )

    if verbose:
        print("GENE INTERSECTION SUMMARY")
        print(f"Excluded datasets : {exclude_keys}")
        print(f"Included datasets : {include_keys}")
        for k in include_keys:
            print(f"  {k:>12} genes: {gene_counts[k]:,}")
        print(f"Common genes (intersection): {len(common_norm):,}")

    # Subset each dataset to the *same ordered* common genes, IN-PLACE
    # Important: avoid a[:, ...].copy() to prevent RAM blow-ups.
    for k in include_keys:
        a = adatas_dict[k]

        # Map common_norm -> original gene names for this dataset
        # (should be all present because intersection was computed from these sets)
        orig_names = gene_maps[k].reindex(common_norm).to_numpy()

        # Safety: ensure no missing (should not happen, but protects against weirdness)
        if np.any(pd.isna(orig_names)):
            missing = [g for g, v in zip(common_norm, orig_names) if pd.isna(v)]
            raise RuntimeError(f"{k}: missing {len(missing)} genes during reindex (unexpected).")

        # Convert to positions in current adata.var_names (fast, no copy of X)
        # Note: orig_names are strings; var_names are current strings.
        idx = a.var_names.get_indexer(orig_names)
        if np.any(idx < 0):
            bad = orig_names[idx < 0]
            raise RuntimeError(f"{k}: could not locate {len(bad)} genes in var_names (unexpected).")

        # In-place subset & reorder
        if inplace:
            # This avoids allocating a full new AnnData matrix
            a._inplace_subset_var(idx)
        else:
            # If you ever want a new object (more memory), you can do this:
            a = a[:, idx].copy()
            adatas_dict[k] = a

        # Standardize var_names if requested (no data copy)
        if gene_norm_mode == "upper":
            a.var_names = pd.Index(common_norm, dtype="object")
        else:
            a.var_names = pd.Index(orig_names, dtype="object")

        a.var_names_make_unique()

        # Record metadata
        a.uns.setdefault("gene_intersection", {})
        a.uns["gene_intersection"].update({
            "excluded_keys": list(exclude_keys),
            "included_keys": list(include_keys),
            "n_common_genes": int(len(common_norm)),
            "gene_case": gene_case,
            "strip": bool(strip),
            "inplace": bool(inplace),
        })

        # (optional) release intermediate arrays promptly
        del orig_names, idx

    return adatas_dict, common_norm


# Usage
exclude_keys = None# ["adata_124395"]
adatas_dict, common_genes = intersect_genes_across_adatas_memory_safe(
    adatas_dict,
    exclude_keys=exclude_keys,
    gene_case="upper",
    strip=True,
    inplace=True,   # <- key: no huge copies
    verbose=True,
)

print(f"Final common genes used: {len(common_genes):,}")
```

## Concatnation + HVG selection

```python
import scanpy as sc
import anndata as ad
import pandas as pd


def build_concat_and_select_hvgs(
    adatas_dict: dict,
    include_keys: list,
    common_genes: list | pd.Index,
    batch_key: str = "dataset",
    n_top_genes: int = 4000,
    flavor: str = "seurat_v3",
    subset: bool = True,
    layer: str | None = None,
    inplace: bool = False,
):
    """
    Concatenate datasets -> reconstruct per-dataset metadata lost by merge='same'
    -> batch-aware HVG selection.

    Parameters
    ----------
    adatas_dict : dict
        Dictionary of AnnData objects keyed by dataset name (e.g. "adata_149614").
    include_keys : list
        Dataset keys to include in concatenation.
    common_genes : list or pd.Index
        Global gene intersection list. All datasets must contain these genes.
    batch_key : str
        Observation column identifying dataset/batch. Default: "dataset".
    n_top_genes : int
        Number of HVGs to select. Default: 4000.
    flavor : str
        HVG selection method. "seurat_v3" supports batch_key. Default: "seurat_v3".
    subset : bool
        If True, returns HVG-subset AnnData as second return value.
        If False, returns None as second value but HVG flags are stored in .var.
    layer : str | None
        Layer to use for HVG computation. None uses adata.X.
    inplace : bool
        If True, modifies objects in adatas_dict directly. Default False.

    Returns
    -------
    adata_all_full : AnnData
        Concatenated AnnData (all common genes) with HVG flags in .var and
        reconstructed metadata columns in .obs.
    adata_all_hvg : AnnData or None
        HVG-subset AnnData if subset=True, else None.

    Notes
    -----
    Metadata reconstruction:
        ad.concat with merge="same" silently drops obs columns that differ
        across datasets (e.g. technology, platform, sequencer), because it
        only preserves columns with identical values everywhere. This function
        rebuilds those columns after concatenation by reading the first value
        from each individual AnnData and mapping it back via the concat_dataset
        label column. The columns reconstructed are:
            technology, platform, sequencer, source_repository
        donor_id survives merge="same" correctly (per-cell, non-conflicting).
    """

    # ── 1. Subset each dataset to common genes ────────────────────────────────
    common_genes = pd.Index(common_genes).astype(str)

    adatas = []
    for k in include_keys:
        a = adatas_dict[k] if inplace else adatas_dict[k].copy()

        if batch_key not in a.obs.columns:
            a.obs[batch_key] = str(k)

        genes_here = pd.Index(a.var_names).astype(str)
        missing = common_genes.difference(genes_here)
        if len(missing) > 0:
            raise ValueError(
                f"[{k}] is missing {len(missing)} common genes (unexpected). "
                f"Example: {missing[:5].tolist()}"
            )

        a = a[:, common_genes].copy()
        a.var_names_make_unique()
        adatas.append(a)

    # ── 2. Concatenate ────────────────────────────────────────────────────────
    # merge="same" preserves only obs columns with identical values across all
    # datasets. Per-dataset columns (technology, platform, sequencer) are
    # silently dropped here — they are recovered in step 3.
    adata_all_full = ad.concat(
        adatas,
        join="outer",
        merge="same",
        label="concat_dataset",
        keys=include_keys,
        fill_value=0,
        index_unique=None,
    )

    # ── 3. Reconstruct per-dataset metadata lost by merge="same" ─────────────
    # Build a lookup: {column_name: {dataset_key: value}}
    # Reads the first obs value from each individual AnnData — correct because
    # technology, platform, sequencer, source_repository are constants per
    # dataset (every cell in the same dataset has the same value).
    _meta_columns = ["technology", "platform", "sequencer", "source_repository"]
    meta_lookup: dict[str, dict[str, str]] = {}

    for k in include_keys:
        a = adatas_dict[k]
        for col in _meta_columns:
            if col in a.obs.columns:
                meta_lookup.setdefault(col, {})[k] = str(a.obs[col].iloc[0])

    # Map reconstructed values back using concat_dataset label as the join key.
    # concat_dataset contains the original include_keys values (e.g. "adata_149614").
    for col, mapping in meta_lookup.items():
        adata_all_full.obs[col] = (
            adata_all_full.obs["concat_dataset"]
            .map(mapping)
            .fillna("Unknown")
        )

    # ── 4. Ensure batch_key column exists and is categorical ─────────────────
    if batch_key not in adata_all_full.obs.columns:
        adata_all_full.obs[batch_key] = adata_all_full.obs["concat_dataset"].astype(str)
    adata_all_full.obs[batch_key] = adata_all_full.obs[batch_key].astype("category")

    # ── 5. Batch-aware HVG selection ──────────────────────────────────────────
    # seurat_v3 flavor is used because it supports batch_key, which prevents
    # dataset-dominant genes from being selected as HVGs.
    sc.pp.highly_variable_genes(
        adata_all_full,
        n_top_genes=n_top_genes,
        flavor=flavor,
        batch_key=batch_key,
        subset=False,
        layer=layer,
        inplace=True,
    )

    # Store HVG selection parameters for reproducibility
    adata_all_full.uns["hvg_params"] = {
        "n_top_genes": int(n_top_genes),
        "flavor": str(flavor),
        "batch_key": str(batch_key),
        "layer_used": layer if layer is not None else "X",
        "feature_space": "gene_intersection_only",
        "n_common_genes": int(adata_all_full.n_vars),
    }

    # ── 6. Subset to HVGs (optional) ─────────────────────────────────────────
    # Subsetting by genes preserves all obs columns including reconstructed
    # metadata, so no additional propagation is needed.
    adata_all_hvg = None
    if subset:
        if "highly_variable" not in adata_all_full.var.columns:
            raise KeyError("HVG selection failed: .var['highly_variable'] missing.")
        hvgs = adata_all_full.var["highly_variable"].values
        adata_all_hvg = adata_all_full[:, hvgs].copy()
        adata_all_hvg.uns["hvg_params"] = adata_all_full.uns["hvg_params"]

    # ── 7. Summary ────────────────────────────────────────────────────────────
    n_hvg = int(adata_all_full.var["highly_variable"].sum())
    print("HVG SELECTION SUMMARY")
    print(f"  concatenated cells : {adata_all_full.n_obs:,}")
    print(f"  common genes used  : {adata_all_full.n_vars:,}")
    print(f"  HVGs selected      : {n_hvg:,}")
    print(f"  batch_key          : {batch_key}")
    print(f"  layer used         : {layer if layer is not None else 'X'}")

    print("\nRECONSTRUCTED METADATA COVERAGE")
    for col in _meta_columns + ["donor_id"]:
        if col in adata_all_full.obs.columns:
            n_unknown = (adata_all_full.obs[col].astype(str) == "Unknown").sum()
            pct_bad = 100 * n_unknown / adata_all_full.n_obs
            n_unique = adata_all_full.obs[col].nunique()
            print(f"  {col}: {pct_bad:.1f}% unknown | {n_unique} unique values")
        else:
            print(f"  {col}: NOT PRESENT")

    return adata_all_full, adata_all_hvg


# ── Call site ─────────────────────────────────────────────────────────────────

include_keys = [
    "adata_149614",
    "adata_151530",
    "adata_125449",
    "adata_115469",
    "adata_140228",
    "adata_146409",
    "adata_138709",
    "adata_136103",
    "adata_124395",
]

adata_all_full, adata_all_hvg = build_concat_and_select_hvgs(
    adatas_dict=adatas_dict,
    include_keys=include_keys,
    common_genes=common_genes,
    batch_key="dataset",
    n_top_genes=4000,
    flavor="seurat_v3",
    subset=True,
    layer=None,  # use adata.X
)

"""
adata_all_full : All 13,181 intersected genes with HVG flag in .var["highly_variable"].
                 Kept for flexibility — allows changing HVG thresholds later or reusing
                 the full gene space for other tasks.

adata_all_hvg  : 4,000 HVG-subset. Smaller, faster matrix used for all integration methods.
                 Contains reconstructed metadata columns:
                     technology, platform, sequencer, source_repository (per-dataset constants)
                     donor_id (per-cell, survived merge="same" correctly)
"""
```

# Save pre-Integration version 

```python
from pathlib import Path
import anndata as ad
import numpy as np
import scipy.sparse as sp
from typing import Optional
import warnings


def save_concatenated_anndata(
    adata_all_full: ad.AnnData,
    adata_all_hvg: Optional[ad.AnnData],
    save_dir: str = "/data1/esraa/Thesis-Project/Data/Processed_data",
    prefix: str = "concatenated",
    compression: str = "gzip",
    save_hvg: bool = True,
    force_sparse: bool = True,
    verbose: bool = True,
) -> dict:
    """
    Save concatenated AnnData objects (full + HVG subset) with proper formatting.
    
    Parameters
    ----------
    adata_all_full : AnnData
        Full concatenated AnnData with all common genes
    adata_all_hvg : AnnData or None
        HVG-subset AnnData (if subset=True was used)
    save_dir : str
        Directory to save files
    prefix : str
        Filename prefix (e.g., "concatenated" -> "concatenated_full.h5ad")
    compression : str
        Compression type: "gzip", "lzf", or None
    save_hvg : bool
        Whether to save the HVG subset (if provided)
    force_sparse : bool
        Convert X to sparse CSR if not already sparse
    verbose : bool
        Print progress messages
    
    Returns
    -------
    dict
        Dictionary with saved file paths and basic counts
    """
    
    save_dir = Path(save_dir)
    save_dir.mkdir(parents=True, exist_ok=True)
    
    results = {
        "full_path": None,
        "hvg_path": None,
        "n_cells": adata_all_full.n_obs,
        "n_genes_full": adata_all_full.n_vars,
        "n_hvgs": None,
    }
    
    # Helper: ensure sparse CSR for memory efficiency
    def _ensure_sparse_csr(adata_obj):
        if force_sparse and not sp.issparse(adata_obj.X):
            if verbose:
                print(f"  Converting X to sparse CSR (was {type(adata_obj.X).__name__})")
            adata_obj.X = sp.csr_matrix(adata_obj.X)
        elif sp.issparse(adata_obj.X) and not isinstance(adata_obj.X, sp.csr_matrix):
            if verbose:
                print(f"  Converting sparse X to CSR (was {type(adata_obj.X).__name__})")
            adata_obj.X = adata_obj.X.tocsr()
        
        # Also handle layers if they exist
        for layer_name in adata_obj.layers.keys():
            layer = adata_obj.layers[layer_name]
            if force_sparse and not sp.issparse(layer):
                adata_obj.layers[layer_name] = sp.csr_matrix(layer)
            elif sp.issparse(layer) and not isinstance(layer, sp.csr_matrix):
                adata_obj.layers[layer_name] = layer.tocsr()
    
    # Helper: ensure obs columns are h5ad-writable
    def _sanitize_obs(adata_obj):
        for col in adata_obj.obs.columns:
            s = adata_obj.obs[col]
            
            # Convert nullable string dtypes to object
            if hasattr(s, 'dtype') and 'string' in str(s.dtype).lower() and s.dtype != object:
                adata_obj.obs[col] = s.astype(object)
            
            # Fix categorical with nullable string categories
            if hasattr(s, 'cat'):
                cats = s.cat.categories
                if hasattr(cats, 'dtype') and 'string' in str(cats.dtype).lower() and cats.dtype != object:
                    adata_obj.obs[col] = s.cat.set_categories(cats.astype(object))
    
    # Save full concatenated object
    if verbose:
        print(f"\n[1/2] Saving full concatenated AnnData...")
        print(f"  Shape: {adata_all_full.n_obs:,} cells × {adata_all_full.n_vars:,} genes")
    
    adata_full_copy = adata_all_full.copy()
    _ensure_sparse_csr(adata_full_copy)
    _sanitize_obs(adata_full_copy)
    
    full_path = save_dir / f"{prefix}_full.h5ad"
    adata_full_copy.write_h5ad(full_path, compression=compression)
    results["full_path"] = str(full_path)
    
    if verbose:
        print(f"  ✓ Saved to: {full_path}")
        print(f"    File size: {full_path.stat().st_size / (1024**3):.2f} GB")
    
    # Save HVG subset if provided
    if save_hvg and adata_all_hvg is not None:
        if verbose:
            print(f"\n[2/2] Saving HVG subset AnnData...")
            print(f"  Shape: {adata_all_hvg.n_obs:,} cells × {adata_all_hvg.n_vars:,} genes")
        
        adata_hvg_copy = adata_all_hvg.copy()
        _ensure_sparse_csr(adata_hvg_copy)
        _sanitize_obs(adata_hvg_copy)
        
        hvg_path = save_dir / f"{prefix}_hvg.h5ad"
        adata_hvg_copy.write_h5ad(hvg_path, compression=compression)
        results["hvg_path"] = str(hvg_path)
        results["n_hvgs"] = adata_hvg_copy.n_vars
        
        if verbose:
            print(f"  ✓ Saved to: {hvg_path}")
            print(f"    File size: {hvg_path.stat().st_size / (1024**3):.2f} GB")
    
    if verbose:
        print(f"\n✓ Save completed")
        print(f"  Total cells: {results['n_cells']:,}")
        print(f"  Common genes (full): {results['n_genes_full']:,}")
        if results['n_hvgs']:
            print(f"  HVGs selected: {results['n_hvgs']:,}")
    
    return results


# Usage after your concatenation code:
results = save_concatenated_anndata(
    adata_all_full=adata_all_full,
    adata_all_hvg=adata_all_hvg,
    save_dir="/data1/esraa/Thesis-Project/Data/Processed_data/post_HVG_intersection",
    prefix="concatenated",
    compression="gzip",
    save_hvg=True,
    force_sparse=True,
    verbose=True,
)  
```
